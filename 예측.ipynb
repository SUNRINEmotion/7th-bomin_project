```python
from keras.models import load_model
import np_utils

from PIL import Image
import numpy as np
```

```python
img = Image.open("test.png").convert("L")
img = np.resize(img, (1,784))
```

```python
test_data = ((np.array(img)/255)-1)* -1
```

```python
model = load_model('predict_model.h5')
```

```python
res = (model.predict(test_data) > 0.5).astype('int32')
```

```python
print(res)
```

```python
def one_hot_to_category(one_hot_encoded):
    # 원핫인코딩된 배열에서 1의 위치를 찾아 인덱스로 변환
    category_index = np.argmax(one_hot_encoded)
    # 예를 들어, 0부터 시작하는 인덱스를 범주형 데이터로 매핑
    category_value = category_index
    return category_value

print(one_hot_to_category(res))
```
